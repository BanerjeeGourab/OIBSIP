```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
df = pd.read_csv('ifood_df.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df[df['MntRegularProds'] < 0]
df['MntRegularProds'] = df['MntRegularProds'].apply(lambda x: 0 if x < 0 else x)
```

```python
df['TotalChildren'] = df['Kidhome'] + df['Teenhome']

spend_cols = ['MntWines', 'MntFruits', 'MntMeatProducts',
              'MntFishProducts', 'MntSweetProducts',
              'MntGoldProds', 'MntRegularProds']

df['TotalSpend'] = df[spend_cols].sum(axis=1)
```

```python
sns.histplot(df['Income'], bins=30, kde=True)
plt.title('Income Distribution')
plt.show()
```

```python
sns.histplot(df['TotalSpend'], bins=30, kde=True)
plt.title('Total Spend Distribution')
plt.show()
```

```python
sns.boxplot(data=df[['Income', 'TotalSpend']])
plt.title('Boxplots for Income and Spend')
plt.show()
```

```python
features = ['Income', 'Recency', 'TotalChildren', 'TotalSpend']
X = df[features]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
inertia = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=df['Cluster'], cmap='viridis')
plt.title('Customer Segments (PCA)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

```python
df.groupby('Cluster')[features].mean().plot(kind='bar', figsize=(10, 6))
plt.title('Average Values per Cluster')
plt.xticks(rotation=0)
plt.show()
```

Insights:
- Cluster 0: High income, high spenders — target with premium products.
- Cluster 1: Low income, low spend — budget-focused deals.
- Cluster 2: Moderate spenders with children — family bundles.
- Cluster 3: High recency (inactive) — re-engagement campaigns.

Recommendations:
- Offer loyalty rewards to Cluster 0.
- Email promotions to inactive Cluster 3.
- Personalized family discounts for Cluster 2.
